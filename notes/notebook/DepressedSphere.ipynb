```python
n = 50
def function(x):
    r, x0 = 0.25, 0
    y = -np.sqrt(r**2 - (x-x0)**2)
    y[np.sqrt((x-x0)**2) > r] = 0
    return y
```

```python
x = linspace(-1,1,100)
y = function(x)
plot(x, y)
```

```python
def grid(nx=10, ny=10):
    
    Gx = range(ny)
    Gy = range(ny)
    x = linspace(-1,1,nx)
    y = function(x)
    for i in range(nx):
        Gx[i] = np.ones(ny)*x[i]
        Gy[i] = [np.r_[np.linspace(-1,y[i],ny/2), np.linspace(y[i],1.,ny/2)]]
    return np.vstack(Gx), np.vstack(Gy)
    # plot(X.flatten(),Y.flatten(), 'b.')
```

```python
X,Y = grid(20,20)
M = Mesh.CurvilinearMesh([X,Y])
```
